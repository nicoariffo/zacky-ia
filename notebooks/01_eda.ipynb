# Exploratory Data Analysis - Zendesk Tickets

Este notebook explora los tickets ingestados para entender:
- Distribución por canal
- Volumen por fecha
- Tags más comunes
- Longitud de tickets
- Calidad de datos

```python
import pandas as pd
from google.cloud import bigquery
from src.config import get_settings

settings = get_settings()
client = bigquery.Client(project=settings.gcp_project_id)
```

```python
# Load raw tickets
query = f"""
SELECT *
FROM `{settings.gcp_project_id}.raw.tickets`
LIMIT 10000
"""
df_raw = client.query(query).to_dataframe()
print(f"Loaded {len(df_raw)} raw tickets")
```

```python
# Basic stats
df_raw.describe()
```

```python
# Channel distribution
df_raw['channel'].value_counts()
```

```python
# Status distribution
df_raw['status'].value_counts()
```
